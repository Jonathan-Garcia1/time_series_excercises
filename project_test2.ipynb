```python
import requests
import pandas as pd

# Replace 'YOUR_APP_TOKEN' with your actual app token
app_token = 'YpcmFqikpOQuHwcbqts0gXAow'

# Define the base API URL
base_url = 'https://data.cityofnewyork.us/resource/h9gi-nx95.json'

# Initialize an empty list to store all data
all_data = []

# Set the initial offset to 0
offset = 0
batch_size = 1000  # Number of records to fetch per request
```

```python

while True:
    # Construct the URL with the app token, date filter, and pagination
    url = f'{base_url}?$where=crash_date between "2023-01-01T00:00:00.000" and "2023-12-31T23:59:59.999"&$$app_token={app_token}&$offset={offset}&$limit={batch_size}'

    # Send an HTTP GET request to the API
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        data = response.json()  # Convert JSON response to Python data
        if len(data) == 0:
            break  # No more data, exit the loop
        all_data.extend(data)  # Add the data to the list
        offset += batch_size  # Increment the offset for the next request
    else:
        print(f"Failed to retrieve data. Status code: {response.status_code}")
        break

# Create a DataFrame using pandas
df = pd.DataFrame(all_data)
```
