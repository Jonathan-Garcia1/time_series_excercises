```python
import requests

# Define the base URL for the NHTSA Crash Data API
base_url = "https://crashviewer.nhtsa.dot.gov/CrashAPI/"

# Define the API endpoint for getting FARS data by year
endpoint = "FARSData/GetFARSData"

# Define the parameters for the API request
params = {
    "dataset": "ACCIDENT",  # Dataset name for accidents
    "FromYear": 2021,       # Start year (2022)
    "ToYear": 2021,         # End year (2022)
    "State": 48,             # State code (You can change this to the desired state)
    "format": "json"        # Data format (JSON)
}
```

```python

# Make the API request
response = requests.get(base_url + endpoint, params=params)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the JSON response
    accident_data = response.json()
    
    # You can now work with the 'accident_data' variable, which contains the data for 2022 accidents.
    
    # For example, you can print the first few records:
    print(accident_data[:5])
    
else:
    print("Failed to retrieve data. Status code:", response.status_code)
```
